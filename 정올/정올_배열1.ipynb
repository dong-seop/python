#### 자가진단 7

```python
nums = list(map(int,input().rstrip().split(" ")))
min_ = 10001
max_ = 0
for i in nums:
    if i< 100:
        if i > max_:
            max_ = i
    else:
        if i < min_:
            min_ = i
if min_ == 10001:
    min_=100
if max_ == 0:
    max_ = 100
print(max_,min_)
```

#### 형성평가 8

```python
nums = list(map(int,input().rstrip().split(" ")))
ind = nums.index(0)
nums = nums[:ind]
nums2 = [i for i in nums if i%5==0]
print("Multiples of 5 : {}".format(len(nums2)))
print("sum : {}".format(sum(nums2)))
print("avg : {:.1f}".format(sum(nums2)/len(nums2)))
```

#### 형성평가 9
